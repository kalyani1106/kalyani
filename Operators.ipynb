Arithmetic

```python
a=10
b=5
s=a+b
d=a-b
m=a*b
div=a/b
mod=a%b
exp=a**b
print(s)
print(d)
print(m)
print(div)
print(mod)
print(exp)
```

Assignment

```python
a=10
print(a)
a+=5
print(a)
a-=5
print(a)
b=5
a*=b
print(a)
a**=b
print(a)
a/=b
print(a)
a%=b
print(a)
```

Comparison

```python
a=10
b=5
print(a==b)
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
print(a!=b)

```

Logical

```python
a=10
print(a>5 and a>8)
print(a>5 or a>15)
print(not(a>8 and a>5))
```

Identity

```python
x=5
y=5
print(x is y)
print(x is not y)
z=2
v=8
print(z is v)
print(z is not v)
```

Membership

```python
a=5
b=8
c=[1,2,5,6,3]
print(a in c)
print(b in c)
print(a not in c)
print(b not in c)
```
